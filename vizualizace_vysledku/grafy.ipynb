## 1. Duration

```python
import matplotlib.pyplot as plt

ulohy = ['U1', 'U2', 'U3']
cas_clovek = [28, 10, 12]
cas_ai = [12, 6, 3]


x = range(len(ulohy))
width = 0.35

barva_clovek = 'cadetblue'  
barva_ai = 'powderblue'      

plt.figure(figsize=(7, 4.5))
bars1 = plt.bar([i - width/2 for i in x], cas_clovek, width, label='Lidský partner', color=barva_clovek)
bars2 = plt.bar([i + width/2 for i in x], cas_ai, width, label='GitHub Copilot', color=barva_ai)

plt.xticks(ticks=x, labels=ulohy, fontsize=11)
plt.ylabel('Čas řešení (min)', fontsize=12)
plt.title('Srovnání času řešení – člověk vs. AI', fontsize=14)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.4)

for bar in bars1 + bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height - 1, str(height),
             ha='center', va='top', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

## 2. Počet iterací

```python
iterace_clovek = [6, 2, 3]
iterace_ai = [3, 1, 1]

plt.figure(figsize=(7, 4.5))
bars3 = plt.bar([i - width/2 for i in x], iterace_clovek, width, label='Lidský partner', color=barva_clovek)
bars4 = plt.bar([i + width/2 for i in x], iterace_ai, width, label='GitHub Copilot', color=barva_ai)

plt.xticks(ticks=x, labels=ulohy, fontsize=11)
plt.ylabel('Počet iterací', fontsize=12)
plt.title('Počet iterací k úspěšnému dokončení – člověk vs. AI', fontsize=14)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.4)

for bar in bars3 + bars4:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height - 0.3, str(height),
             ha='center', va='top', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

## 3. Kvalita kódu

```python
import matplotlib.pyplot as plt

# Data
ulohy = ['U1', 'U2', 'U3']
srozumitelnost_clovek = [4, 4, 4]
technicka_clovek = [3, 4, 3]
srozumitelnost_ai = [5, 5, 5]
technicka_ai = [4, 5, 4]

x = range(len(ulohy))
width = 0.35

# Barvy
sroz_clovek_color = 'powderblue'  # světle modrá
tech_clovek_color = 'cadetblue'  # tmavě modrá
sroz_ai_color = '#e0bbff'      # světle orchyd
tech_ai_color = '#a678b6'      # tmavě orchyd

# Vykreslení
plt.figure(figsize=(9, 5))

# Lidský partner
bars_c1 = plt.bar([i - width/2 for i in x], srozumitelnost_clovek, width,
                  label='Srozumitelnost – člověk', color=sroz_clovek_color)
bars_c2 = plt.bar([i - width/2 for i in x], technicka_clovek, width,
                  bottom=srozumitelnost_clovek, label='Technická kvalita – člověk', color=tech_clovek_color)

# GitHub Copilot
bars_a1 = plt.bar([i + width/2 for i in x], srozumitelnost_ai, width,
                  label='Srozumitelnost – AI', color=sroz_ai_color)
bars_a2 = plt.bar([i + width/2 for i in x], technicka_ai, width,
                  bottom=srozumitelnost_ai, label='Technická kvalita – AI', color=tech_ai_color)

# Osy a titulky
plt.xticks(ticks=x, labels=ulohy, fontsize=11)
plt.ylim(0, 11)
plt.ylabel('Hodnocení (0–10)', fontsize=12)
plt.title('Hodnocení kvality kódu – Lidský partner vs. AI (stacked)', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Popisky celkového skóre nad sloupcem
for i in x:
    total_clovek = srozumitelnost_clovek[i] + technicka_clovek[i]
    total_ai = srozumitelnost_ai[i] + technicka_ai[i]
    plt.text(i - width/2, total_clovek + 0.2, str(total_clovek),
             ha='center', va='bottom', fontsize=10)
    plt.text(i + width/2, total_ai + 0.2, str(total_ai),
             ha='center', va='bottom', fontsize=10)

# Legenda
plt.legend(ncol=2, loc='upper center', bbox_to_anchor=(0.5, -0.15))

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
ulohy = ['U1', 'U2', 'U3']
x = range(len(ulohy))
width = 0.35

# --- Výkonové metriky ---
cas_clovek = [28, 10, 12]
cas_ai = [12, 6, 3]
iterace_clovek = [6, 2, 3]
iterace_ai = [3, 1, 1]

# --- Čmuchař ---
srozumitelnost_clovek = [4, 4, 4]
technicka_clovek = [3, 4, 3]
srozumitelnost_ai = [5, 5, 5]
technicka_ai = [4, 5, 4]

# --- Barvy ---
barva_clovek = 'cadetblue'
barva_ai = 'powderblue'
sroz_ai_color = '#e0bbff'
tech_ai_color = '#a678b6'

# --- Vykreslení všech grafů pod sebou ---
fig, axs = plt.subplots(3, 1, figsize=(7, 8))

# --- Graf 1: Čas ---
bars1 = axs[0].bar([i - width/2 for i in x], cas_clovek, width, label='Lidský partner', color=barva_clovek)
bars2 = axs[0].bar([i + width/2 for i in x], cas_ai, width, label='GitHub Copilot', color="#a678b6")
axs[0].set_title('Srovnání času řešení – člověk vs. AI', fontsize=12, fontweight='bold')
axs[0].set_ylabel('Čas (min)', fontsize=11)
axs[0].set_xticks(x)
axs[0].set_xticklabels(ulohy)
axs[0].grid(axis='y', linestyle='--', alpha=0.4)
axs[0].legend(fontsize=9)
for bar in bars1 + bars2:
    height = bar.get_height()
    axs[0].text(bar.get_x() + bar.get_width()/2, height - 1, str(height),
                ha='center', va='top', fontsize=9, color='black')

# --- Graf 2: Iterace ---
bars3 = axs[1].bar([i - width/2 for i in x], iterace_clovek, width, label='Lidský partner', color=barva_clovek)
bars4 = axs[1].bar([i + width/2 for i in x], iterace_ai, width, label='GitHub Copilot', color="#a678b6")
axs[1].set_title('Počet iterací k úspěšnému dokončení', fontsize=12, fontweight='bold')
axs[1].set_ylabel('Iterace', fontsize=11)
axs[1].set_xticks(x)
axs[1].set_xticklabels(ulohy)
axs[1].legend(fontsize=9)
axs[1].grid(axis='y', linestyle='--', alpha=0.4)
for bar in bars3 + bars4:
    height = bar.get_height()
    axs[1].text(bar.get_x() + bar.get_width()/2, height - 0.3, str(height),
                ha='center', va='top', fontsize=9, color='black')



# --- Graf 3: Hodnocení kvality kódu (stacked, 2 barvy) ---
bars_c1 = axs[2].bar([i - width/2 for i in x], srozumitelnost_clovek, width,
                     color='powderblue')
bars_c2 = axs[2].bar([i - width/2 for i in x], technicka_clovek, width,
                     bottom=srozumitelnost_clovek, label='Lidský partner', color='cadetblue')

bars_a1 = axs[2].bar([i + width/2 for i in x], srozumitelnost_ai, width,
                     color='#e0bbff')
bars_a2 = axs[2].bar([i + width/2 for i in x], technicka_ai, width,
                     bottom=srozumitelnost_ai, label='GitHub Copilot', color='#a678b6')

axs[2].set_title('Hodnocení kvality kódu', fontsize=12, fontweight='bold')
axs[2].set_ylabel('Skóre (0–10)', fontsize=11)
axs[2].set_xticks(x)
axs[2].set_xticklabels(ulohy)
axs[2].grid(axis='y', linestyle='--', alpha=0.4)
axs[2].set_ylim(0, 11)

# Legenda: jen dvě položky (člověk vs AI)
axs[2].legend(fontsize=9, loc="lower center", ncol=2)

# Popisky celkových hodnot
for i in x:
    total_clovek = srozumitelnost_clovek[i] + technicka_clovek[i]
    total_ai = srozumitelnost_ai[i] + technicka_ai[i]
    axs[2].text(i - width/2, total_clovek + 0.2, str(total_clovek),
                ha='center', va='bottom', fontsize=9)
    axs[2].text(i + width/2, total_ai + 0.2, str(total_ai),
                ha='center', va='bottom', fontsize=9)


plt.tight_layout()
plt.show()
```

## Introspekce

```python
import matplotlib.pyplot as plt

# Metriky a kategorie
metriky = ['Úzkost (PAS)', 'Motivace (IMI)', 'Spolupráce (CLS)']
kategorie = ['Logické úlohy', 'Úlohy s daty', 'Algoritmické úlohy']
x = range(len(metriky))
width = 0.35

# Data
clovek = [
    [3.4, 4.5, 4.8],  # Logické
    [3.1, 4.3, 4.5],  # Datové
    [3.3, 4.6, 4.9]   # Algoritmické
]
ai = [
    [2.2, 4.2, 3.7],  # Logické
    [1.9, 4.0, 3.5],  # Datové
    [2.1, 4.1, 3.6]   # Algoritmické
]

# Barvy
barva_clovek = 'cadetblue'
barva_ai = '#b184c4'

# Vykreslení grafů
fig, axs = plt.subplots(3, 1, figsize=(7, 7))

for i in range(3):
    bars1 = axs[i].bar([p - width/2 for p in x], clovek[i], width,
                       label='Lidský partner', color=barva_clovek)
    bars2 = axs[i].bar([p + width/2 for p in x], ai[i], width,
                       label='GitHub Copilot', color=barva_ai)

    axs[i].set_title(f'{kategorie[i]}', fontsize=12, fontweight='bold')
    axs[i].set_xticks(x)
    axs[i].set_xticklabels(metriky, fontsize=10)
    axs[i].set_ylim(0, 5.2)
    axs[i].set_ylabel('Skóre (1–5)', fontsize=10)
    axs[i].grid(axis='y', linestyle='--', alpha=0.4)

    # Hodnoty ve sloupcích
    for bar in bars1 + bars2:
        height = bar.get_height()
        axs[i].text(bar.get_x() + bar.get_width()/2, height - 0.15,
                    f'{height:.1f}', ha='center', va='top', fontsize=9, color='black')

# Legenda nad grafy (nezasahuje do názvů)
fig.legend(['Lidský partner', 'GitHub Copilot'], loc='upper center', ncol=2, bbox_to_anchor=(0.5, 1.04), fontsize=9)

plt.tight_layout()
plt.show()
```
